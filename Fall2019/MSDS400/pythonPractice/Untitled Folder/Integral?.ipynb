```python
import matplotlib.pyplot as plt
from numpy import poly1d, linspace
```

```python
x = linspace(-4, 4, 101)
y = 1/x 
```

```python
plt.figure()
plt.plot(x, y, label='y = 1/x')
plt.legend(loc='best')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.ylim(0,10)
plt.xlim(0,10)
plt.vlines(x=4, ymin=0, ymax=10, color='orange')
plt.vlines(x=2, ymin=0, ymax=10, color='lime')
plt.grid()
plt.fill_between([2,4],[1/2,1/4], color='grey',alpha=0.2)
plt.title('y = 1/x Bounded by x = 2 and x = 4')
plt.show()
```

```python
plt.figure()
plt.plot(x, y, label='y = 1/x')
plt.legend(loc='best')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.fill_between(x, 0, y, color='grey',alpha=0.2)
plt.title('y = 1/x Unbounded')
plt.show()
```
