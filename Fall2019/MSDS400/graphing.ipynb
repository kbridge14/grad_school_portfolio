```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y1(x):
    return 1/x

def y2(x):
    return x**2 - 7

def y3(x):
    return 6*x**4

def y4(x):
    return x**3 - 2*x**2 + 3*x + 40

x = np.array(range(-1000, 1000)) 
```

```python
plt.plot(x, y1(x))
```

```python
plt.plot(x, y2(x))
```

```python
plt.plot(x, y3(x))
```

```python
plt.plot(x, y4(x))
```
